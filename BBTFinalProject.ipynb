```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import roc_curve, auc
```

```python
# address missingness
# dropping rows with missing occupation/country data
df2 = df2[df2['occupation'].notna() & df2['native-country'].notna()]

# impute with means
df2['age_na'] = df2['age'].isna()
df2['hours_na'] = df2['hours-per-week'].isna()
df2['age'].fillna(value=df2['age'].mean(), inplace=True)
df2['hours-per-week'].fillna(value=df2['hours-per-week'].mean(), inplace=True)

np.sum(df2.isna())
```

```python
# binary encoding: marital status (has been married/not), native country (US native/not)
df2['wasMarried'] = np.where(df2['marital-status'] == 'Never-married', 0, 1)
df2['usNative'] = np.where(df2['native-country'] == 'United-States', 1, 0)
df2['female'] = np.where(df2['sex_selfID'] == 'Female', 1, 0)
print(df2['wasMarried'].unique())
print(df2['usNative'].unique())

# one-hot encoding: occupation, race
df_occ = pd.get_dummies(df2['occupation'], prefix='job_')
df_race = pd.get_dummies(df2['race'], prefix='race_')
df3 = df2.join(df_occ).join(df_race)

df3.head()
```

```python
# recode education
lt_bach = ['Preschool', '1st-4th', '5th-6th', '7th-8th', '9th', '10th', '11th', '12th', 
           'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Some-college']
df3['new-edu'] = np.where([edu in lt_bach for edu in df3['education']], "< Bachelors", ">= Bachelors")
df3 = df3.drop(columns="education")
df3['new-edu'].unique()
```

```python
# drop unnecessary columns
cols = ['marital-status', 'native-country', 'sex_selfID', 'occupation', 'race', 'age_na', 'hours_na']
df3 = df3.drop(columns=cols)
df3.head()
```

```python
# split dataset
X = df3.drop(columns='new-edu')
y = df3['new-edu']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)
```

```python
# initial GBDT model
model = GradientBoostingClassifier(random_state=123)
model.fit(X_train, y_train)
gbdt_predictions = model.predict_proba(X_test)[:,-1]

# initial AUC value
fpr, tpr, thresholds = roc_curve(y_test, gbdt_predictions, pos_label=">= Bachelors")
auc(fpr, tpr)
```

```python
# set parameters for grid search
param_grid = {"n_estimators": range(50,101,10), 
              "min_samples_leaf": range(50,101,10), 
              "max_depth": [2**i for i in range(4)]}

# run grid search
new_model = GradientBoostingClassifier(random_state=123)
grid = GridSearchCV(new_model, param_grid, scoring='roc_auc', verbose=3, cv=5)
grid_search = grid.fit(X_train, y_train)

best_params = grid_search.best_params_
best_params
```

```python
# optimized GBDT model
best_model = GradientBoostingClassifier(random_state=123, 
                                        n_estimators=best_params['n_estimators'], 
                                        max_depth=best_params['max_depth'], 
                                        min_samples_leaf=best_params['min_samples_leaf'])
best_model.fit(X_train, y_train)
best_predictions = best_model.predict_proba(X_test)[:,-1]

# optimized AUC value
fpr2, tpr2, thresholds2 = roc_curve(y_test, best_predictions, pos_label=">= Bachelors")
auc(fpr2, tpr2)
```

```python
# saving best model
import pickle
pickle.dump(best_model, open('census_gbdtmodel.pkl', 'wb'))
```
